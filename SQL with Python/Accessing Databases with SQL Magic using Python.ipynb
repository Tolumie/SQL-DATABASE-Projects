#### Using Python Variables in your SQL Statements

##### You can use python variables in your SQL statements by adding a ":" prefix to your python variable names.

##### For example, if I have a python variable `country` with a value of `"Canada"`, I can use this variable in a SQL query to find all the rows of students from Canada.


```python
country = "Canada"
%sql select * from INTERNATIONAL_STUDENT_TEST_SCORES where country = :country
```

#### Assigning the Results of Queries to Python Variables


##### You can use the normal python assignment syntax to assign the results of your queries to python variables.

##### For example, I have a SQL query to retrieve the distribution of test scores (i.e. how many students got each score). I can assign the result of this query to the variable `test_score_distribution` using the `=` operator.


```python
test_score_distribution = %sql SELECT test_score as "Test_Score", count(*) as "Frequency" from INTERNATIONAL_STUDENT_TEST_SCORES GROUP BY test_score;
test_score_distribution
```

#### Converting Query Results to DataFrames


##### You can easily convert a SQL query result to a pandas dataframe using the `DataFrame()` method. Dataframe objects are much more versatile than SQL query result objects. For example, we can easily graph our test score distribution after converting to a dataframe.


```python
dataframe = test_score_distribution.DataFrame()

%matplotlib inline
# uncomment the following line if you get an module error saying seaborn not found
# !pip install seaborn==0.9.0
import seaborn

plot = seaborn.barplot(x='Test_Score',y='Frequency', data=dataframe)
```

--- 

Trigger that logs a message every time a new record is inserted into the INTERNATIONAL_STUDENT_TEST_SCORES table.

```python
%%sql

-- Log whenever a new student is added
CREATE TABLE Student_Log (
    log_message TEXT,
    log_time TIMESTAMP DEFAULT CURRENT_TIMESTAMP
);

CREATE TRIGGER Log_Student_Add
AFTER INSERT ON INTERNATIONAL_STUDENT_TEST_SCORES
BEGIN
    INSERT INTO Student_Log (log_message)
    VALUES ('Student added: ' || NEW.first_name || ' ' || NEW.last_name);
END;
```

```python
%%sql

-- View the Structure of the New Table
PRAGMA table_info('Student_Log');
```

```python
# Testing the Trigger
```

```python
%%sql
-- Insert a new student and check the log
INSERT INTO INTERNATIONAL_STUDENT_TEST_SCORES (country, first_name, last_name, test_score)
VALUES ('France', 'Jean', 'Dupont', 60);

-- Query the log to confirm the trigger worked
SELECT * FROM Student_Log;
```

```python
# Bar Plot for Test Scores by Country WITH percentage displayed (first five)
```

```python
# Execute the SQL query to get data
result = %sql SELECT country, AVG(test_score) AS avg_score FROM INTERNATIONAL_STUDENT_TEST_SCORES GROUP BY country LIMIT 5;

# Convert the result to a pandas DataFrame
import pandas as pd
df = pd.DataFrame(result)

# Plotting the data using Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the percentage of each average score
df['percentage'] = (df['avg_score'] / df['avg_score'].sum()) * 100

# Plot the bar chart
sns.barplot(x='country', y='avg_score', data=df, palette='viridis')

# Annotate each bar with its percentage
for i, row in df.iterrows():
    plt.text(i, row['avg_score'], f"{row['percentage']:.1f}%", ha='center', va='bottom', fontsize=10)

# Customize the plot
plt.xticks(rotation=45, ha='right')
plt.title('Average Test Scores by Country')
plt.xlabel('Country')
plt.ylabel('Average Test Score')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Box Plot for Test Scores Distribution by Country
result = %sql SELECT country, test_score FROM INTERNATIONAL_STUDENT_TEST_SCORES limit 20 ;

# Convert the result to a pandas DataFrame
df = pd.DataFrame(result)

# Plotting the data using Seaborn
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='country', y='test_score', data=df)
plt.xticks(rotation=45, ha='right')
plt.title('Test Scores Distribution by Country (Limited to 5 Values)')
plt.xlabel('Country')
plt.ylabel('Test Score')
plt.tight_layout()
plt.show()
```
